```python
# The current directory is selected 
parent_directory = '/Users/vinipatel/Desktop/lab2/CDR3 loops'

# A directory is created for the storage of the loops
CDR3a_dir = 'CDR3a_loops' 
CDR3a_path = os.path.join(parent_directory, CDR3a_dir)
os.mkdir(CDR3a_path)


# the names of the pdb files are stored in an excel sheet along with the starting residue number and the ending resifue number of D and E chain
import openpyxl as xl
import os
file_name = []
wb = xl.load_workbook('CDR3 loops.xlsx', data_only = True)
sheet = wb['Sheet1']

# the variables in the sheet are extracted and read
for row in range(2, sheet.max_row +1):
    cell2 = sheet.cell(row, 2)
    cell3 = sheet.cell(row, 3)
    cell8 = sheet.cell(row, 8)
    cell7 = sheet.cell(row, 7)
    cell9 = sheet.cell(row, 9)
    cell10 = sheet.cell(row, 10)
    cell11 = sheet.cell(row, 11)
    cell12 = sheet.cell(row, 12)
    cell13 = sheet.cell(row, 13)
    cell14 = sheet.cell(row, 14)
    file_name = cell2.value
    chain_D = cell7.value
    chain_E = cell11.value
    Start_res_num_D =cell8.value
    end_res_num_D = cell9.value
    Start_res_num_E =cell12.value
    end_res_num_E = cell13.value
    
# PDB file with the TCR-peptide-MHC molecule is opened
    input_file = file_name + '_mod.pdb'
    with open(input_file , 'r') as pdb_file:
        lines = pdb_file.readlines()
     
# the CDR3 loops are selected 
    for i , line in enumerate(lines):
        if line[21:22] == chain_D and line[23:26] == ' '* (3-len(str(Start_res_num_D))) + str(Start_res_num_D) and line[13:15] == 'N ' and  line[17:20] == 'CYS':
            fix_begin = i   
        if line[21:22] == chain_D and line[23:26] == str(end_res_num_D+1) and line[13:15] == 'N ' :
            fix_end = i 

# The CDR3 loops are extracted
    CDR3a_loop = lines[fix_begin:fix_end]
    CDR3a_loop +=['TER\n']


# The atoms are renumbered starting from 1
    b = 0
    for i ,line in enumerate(CDR3a_loop):
        if line[0:4] == 'ATOM':
            b+=1
            a = str(b)
        #print(a)
            if len(a) == 1: 
                c = ' ' + ' ' + ' '+ a    
            if len(a) == 2: 
                c = ' ' + ' ' + a
            if len(a) == 3: 
                c = ' ' + a
            if len(a) == 4: 
                c = a
        ##print(a)   
            CDR3a_loop[i] = CDR3a_loop[i][:7] + c + CDR3a_loop[i][11:] 
  
# The PDB file with the extracted loop is saved into the specific directory
    output_file = file_name + '_CDR3a_loop.pdb'
    with open(output_file ,'w') as f:
         for line in CDR3a_loop:
              f.write(line)
  

  
```
