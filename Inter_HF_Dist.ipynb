```python
import os
import random
import shutil
import pandas as pd
import openpyxl as xl
import numpy as np
import math
os.chdir (r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/1ao7CDR3a_vary_loop/')
```

```python
# Hydrophobic interactions between the CDR3 alpha loop and CDR3 beta loop as well as hydrophobic interactions between the CDR3 alpha chain and peptide and CDR3 beta chain and peptide
# C chain is  peptide chain
def category(dist):
    """
    The category of distance.
    """
    if dist < 0.5:
        cate = "<0.5"
    elif dist >= 0.5 and dist < 1:
        cate = "[0.5, 1]"
    elif dist >= 1 and dist < 1.5:
        cate = "[1, 1.5]"
    elif dist >= 1.5 and dist < 3:
        cate = "[1.5, 3]"
    elif dist >= 3 and dist < 5:
        cate = "[3, 5]"
    else:
        cate = ">=5"
    return cate

def distance_calculate2(pdb_name, lines, file):
    """
    Calculate the distance between atoms.
    """
    carbons = ["CA ","CB ","CG ","CD ","CE ","CZ ","CH ","CD1","CD2","CG1","CG2","CE1","CE2","CE3","CZ1","CZ2","CH1","CH2", "SD", "SG"]
    atom_name = []
    chain_name = []
    res_name = []
    res_num = []
    x_coord = []
    y_coord =[]
    z_coord =[]
    for line in lines:
        if line[13:16] in carbons:
            atom_name.append(str(line[13:16]))
            chain_name.append(str(line[21:22]))
            res_name.append(str(line[17:20]))
            res_num.append(int(line[23:26]))
            x_coord.append(float(line[31:38]))
            y_coord.append(float(line[39:46]))
            z_coord.append(float(line[47:54]))

    output = []
    total_HF_pep_D = 0
    total_HF_D_E = 0
    # Added column: peptide name
    file_name = file
    for i in range(0, len(x_coord), 1):
        for j in range(i+1, len(x_coord), 1):
            if chain_name [i] == "C" and chain_name[j] == "D":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = round(math.sqrt(xx+yy+zz),2)
                    if dist < 10:
                        cate = category(dist)
                        output.append([ file_name, atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
                    else:
                        StopIteration
            if chain_name [i] == "D" and chain_name[j] == "E":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = round(math.sqrt(xx+yy+zz),2)
                    if dist < 10:
                        cate = category(dist)
                        output.append([ file_name, atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
                    else:
                        StopIteration
            if chain_name [i] == "C" and chain_name[j] == "E":
                if res_num[i] != res_num[j] or res_num[i] == res_num[j]:
                    xx = (x_coord[i]-x_coord[j])**2
                    yy = (y_coord[i]-y_coord[j])**2
                    zz = (z_coord[i]-z_coord[j])**2
                    dist = round(math.sqrt(xx+yy+zz),2)
                    if dist < 10:
                        cate = category(dist)
                        output.append([ file_name, atom_name[i], chain_name[i], res_name[i], res_num[i], atom_name[j], chain_name[j], res_name[j], res_num[j], float(dist), cate])
                    else:
                        StopIteration
            
    output = np.array(output)
    # print(output)
    
    return output
```

```python
wb = xl.load_workbook('/Users/vinipatel/Desktop/lab2/CDR3_loops/AddProt_Master.xlsx', data_only = True)
sheet = wb['Sheet4']
# using for loop- assign variable to each cells of different columns
for row in range(2, sheet.max_row + 1):    
    cell1 = sheet.cell(row, 1)    
    cell2 = sheet.cell(row, 2)    
    pdb_names = str(cell1.value)
    conformations_l = cell2.value

    res_all = []
    conformations = []
    conformations_list = conformations_l.split()      
    conformations.append(conformations_list)
    file_names = str(pdb_names)

    HF_dist_file = r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/' +file_names+ 'CDR3a_HF_dist'
    new_path = os.path.join(r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/', HF_dist_file)
    os.mkdir(new_path)
    print(HF_dist_file)
      
    CDR3_file = r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/' +file_names+ 'CDR3a_vary_loop'
    os.chdir(CDR3_file)
    print(CDR3_file)
    imp_file = []
    files = os.listdir(CDR3_file)
    for file in files:
        if file.endswith('.pdb'):
            imp_file.append(file)
        
    file_name3 = file_names + "_CDR3a_distanceAP.xlsx"
    with pd.ExcelWriter(file_name3) as writer:
        for file2, conf_num in zip(imp_file,conformations_list):
            # print(conf_num)
            if file2.endswith('.pdb'):
                # print(file)
                src = r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/' + file_names + 'CDR3a_vary_loop/' + file_names +'_CDR3a_mod_vary'+ conf_num+ 'new.pdb'
                print(src)
                pdb_name = src
                input_file = src 
                with open(input_file,'r') as pdb_file:
                        lines = pdb_file.readlines()
                # file_name3 = file + "_CDR3b_distanceAP.xlsx"
                # file_name4 = file + "_CDR3b_distance_final.xlsx"
                file_name = file_names +'_' + conf_num
                print(file_name)
                res = distance_calculate2(pdb_name, lines, file_name)
                columns = [ "file_name","atom name i", "chain name i", "res name i", "res num i", "atom name j", "chain name j", "res name j", "res num j", "distance", "distance category"] 
    
                pd.DataFrame(res, columns=columns).to_excel(writer, sheet_name=file_name, index=None)
    # Complilation of the hydrophobic interactions for all the conformations of CDR3 alpha conformations
    shutil.move(r'/Users/vinipatel/Desktop/lab2/CDR3_loops/CDR3a_loop_vary/' +file_names + 'CDR3a_vary_loop/' + file_name3, HF_dist_file)
    os.chdir(HF_dist_file)
    excel_file = pd.read_excel(file_name3,sheet_name=None)
    file_name2 = file_names + "_CDR3a_distanceAP" + "_compile.xlsx"
    df2 = pd.concat(excel_file.values())
    df2.to_excel(file_name2,index=None)
```
