```python
# The code can be utilized for both data files containing CDR3 alpha loop variation and CDR3 beta loop variation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl as xl
import shutil
import os
from scipy.stats import mannwhitneyu
from matplotlib.patches import Patch
```

```python
# Reading the excel file
df = pd.read_excel('/Users/vinipatel/Desktop/lab2/CDR3_loops/Results/Solvate_CDR3a/compilation/solvate_cdr3a.xlsx')
```

```python
# Selecting data based on the categories
structures_x_ray = df.loc[df['Binding'] == 0]
structures_pep_vary = df.loc[df ['Binding'] == 1]
structures_CDR3a_vary = df.loc[df ['Binding'] == 2]
structures_CDR3a_pep_vary = df.loc[df['Binding'] == 3]
```

```python
# Creating excel sheets containting 
os.chdir('/Users/vinipatel/Desktop/lab2/CDR3_loops/Results/Solvate_CDR3a/compilation/')
structures_x_ray.to_excel('structures_x_ray.xlsx', index=False)
structures_pep_vary.to_excel('structures_pep_vary.xlsx', index=False)
structures_CDR3a_vary.to_excel('structures_CDR3a_vary.xlsx', index=False)
structures_CDR3a_pep_vary.to_excel('structures_CDR3a_pep_vary.xlsx', index=False)
```

Prot-Lig HB- Hydrogen Bonding Interactions
Statistical Tests and Plot Generation Code

```python
statistic, p_value = mannwhitneyu(structures_x_ray['Prot-Lig HB'], structures_pep_vary['Prot-Lig HB'])
print(p_value)
```

```python
statistic2, p_value2 = mannwhitneyu(structures_x_ray['Prot-Lig HB'],structures_CDR3a_vary['Prot-Lig HB'])
print(p_value2)
```

```python
statistic3, p_value3 = mannwhitneyu(structures_x_ray['Prot-Lig HB'], structures_CDR3a_pep_vary['Prot-Lig HB'])
print(p_value3)
```

```python
# Generate the plots with the above calculated p - values
# Read the excel file
df = pd.read_excel('/Users/vinipatel/Desktop/lab2/CDR3_loops/Results/Solvate_CDR3a/compilation/solvate_cdr3a.xlsx')
sns.set_style("white")
sns.set_context("poster")
# Generate the Box-plot
axes = sns.boxplot(data=df, x='Binding', y='Prot-Lig HB',fliersize=0, zorder=5,)
# Generate the significant bars to add the p - value
x1, x2 = 0, 1  
y, h, col = df['Prot-Lig HB'].max() + 5, 5, 'k'

plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
plt.text((x1+x2)*.5, y+h, f' {p_value:.3f}', ha='center', va='bottom', color=col, fontsize = 18)

x1, x3 = 0, 2  
y2, h2, col2 = df['Prot-Lig HB'].max() + 10, 10, 'k'

plt.plot([x1, x1, x3, x3], [y2, y2+h2, y2+h2, y2], lw=1.5, c=col2)
plt.text((x1+x3)*.5, y2+h2, f' {p_value2:.3f}', ha='center', va='bottom', color=col2, fontsize = 18)
x1, x4 = 0, 3  
y3, h3, col3 = df['Prot-Lig HB'].max() + 15, 15, 'k'

plt.plot([x1, x1, x4, x4], [y3, y3+h3, y3+h3, y3], lw=1.5, c=col3)
plt.text((x1+x4)*.5, y3+h3, f' {p_value3:.3f}', ha='center', va='bottom', color=col3,fontsize = 18)


x1 = [0,1,2,3]
# Add the x -axis labes
squad = ['Binding', 'NB'"\n" '(pep vary)','NB'"\n" '(CDR3a vary)', 'NB'"\n" '(CDR3a'"\n" '+ '"\n" 'pep vary)']
axes.set_xticks(x1)
axes.set_xticklabels(squad, minor=False, rotation=0, fontsize = 16)

plt.xlabel(' ')
# Add y-axis label
plt.ylabel('Hydrogen Bonding Interactions', fontsize = 18)
sns.despine()
# Save the Plots
plt.savefig("/Users/vinipatel/Desktop/lab2/CDR3_loops/Results/Solvate_CDR3a/compilation/Solvate_HB.jpeg", bbox_inches='tight')
```
